```python
import requests
import pandas as pd
import time
import warnings

# Suppress the FutureWarning for the append method
warnings.filterwarnings("ignore", category=FutureWarning)

# Define the base URL for the World Bank API
base_url = "http://api.worldbank.org/v2/country/{}/indicator/{}?format=json&date=2010:2023&per_page=100"

# Define the series codes
series_codes = {
    "VC.IDP.NWDS": "Internally displaced persons, new displacement associated with disasters (number of cases)",
    "FS.AST.PRVT.GD.ZS": "Domestic credit to private sector (% of GDP)",
    "FR.INR.LNDP": "Interest rate spread (lending rate minus deposit rate, %)",
    "GB.XPD.RSDV.GD.ZS": "Research and development expenditure (% of GDP)",
    "SP.POP.SCIE.RD.P6": "Researchers in R&D (per million people)",
    "GC.DOD.TOTL.GD.ZS": "Central government debt, total (% of GDP)",
    "EG.FEC.RNEW.ZS": "Renewable energy consumption (% of total final energy consumption)",
    "LP.LPI.ITRN.XQ": "Logistics performance index: Ease of arranging competitively priced shipments (1=low to 5=high)", # Desprod
    "LP.LPI.TIME.XQ": "Logistics performance index: Frequency with which shipments reach consignee within scheduled or expected time, score (1=low to 5=high)", # Desprod
    "LP.LPI.INFR.XQ": "Logistics performance index: Quality of trade and transport-related infrastructure (1=low to 5=high)", # Infra # Desprod 
    "LP.LPI.TRAC.XQ":"Logistics performance index: Ability to track and trace consignments, score (1=low to 5=high)", # Desprod
    "LP.LPI.CUST.XQ":"Logistics performance index: Efficiency of the clearance process, score (1=low to 5=high)", # Desprod
    "LP.LPI.LOGS.XQ":"Logistics performance index: Competence and quality of logistics services, score (1=low to 5=high)", # Infra  # Desprod
    "IE.PPN.TRAN.CD": "Public private partnerships investment in transport (current US$)",
    "SH.STA.BASS.ZS": "People using at least basic sanitation services (% of population)",
    "SH.H2O.BASW.ZS": "People using at least basic drinking water services (% of population)",
    "SE.PRM.NENR": "School enrollment, primary (% net)",
    "SE.PRE.ENRR": "School enrollment, preprimary (% gross)",
    "SE.SEC.NENR": "School enrollment, secondary (% net)",
    "SE.SEC.CUAT.UP.ZS": "Educational attainment, at least completed upper secondary, population 25+, total (%) (cumulative)",
    "SL.GDP.PCAP.EM.KD": "GDP per person employed (constant 2021 PPP $)",
    "VC.IHR.PSRC.P5": "Intentional homicides (per 100,000 people)",
    "SI.POV.DDAY":"Poverty headcount ratio at $2.15 a day (2017 PPP) (% of population)",
    "SI.POV.LMIC":"Poverty headcount ratio at $3.65 a day (2017 PPP) (% of population)",
    "SI.POV.UMIC":"Poverty headcount ratio at $6.85 a day (2017 PPP) (% of population)"    
}

# Function to fetch data for a specific indicator and country
def fetch_data(country_code, indicator_code):
    url = base_url.format(country_code, indicator_code)
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return data[1] if len(data) > 1 else []
    else:
        print(f"Error fetching data for {country_code} - {indicator_code}: {response.status_code}")
        return []

# Create an empty DataFrame to store the results
wdi_df_raw = pd.DataFrame()

# Split the country_codes dictionary into batches of 5 countries
country_codes_list = list(country_codes.items())
batch_size = 5

for i in range(0, len(country_codes_list), batch_size):
    batch = country_codes_list[i:i + batch_size]
    
    for country_name, country_code in batch:
        for series_code, series_name in series_codes.items():
            print(f"Downloading data for {country_name} - {series_name}")
            records = fetch_data(country_code, series_code)
            for record in records:
                wdi_df_raw = wdi_df_raw.append({
                    "country": country_name,
                    "country_code": country_code,
                    "year": record['date'],
                    "indicator": series_name,
                    "indicator_code": series_code,
                    "value": record['value']
                }, ignore_index=True)
    
    # Save the DataFrame after each batch
    wdi_df_raw.to_csv("wdi_data_raw.csv", index=False)
    print(f"Data saved after processing batch {i//batch_size + 1}")
    
    # Print remaining ISO country codes
    remaining_countries = [code for _, code in country_codes_list[i + batch_size:]]
    print("Remaining countries ISO codes:", remaining_countries)
    
    # Pause for 3 seconds between each batch
    print("Pausing for 3 seconds...")
    time.sleep(3)

# Display the DataFrame structure and the first few rows
print(wdi_df_raw.head())
```

##### **WHO GHO Api**

Because this API is a bit cumbersome, we first make sure that we have the current codes for each indicator. 


```python
import requests

# Fetch indicator list from WHO GHO API
indicators_url = "https://ghoapi.azureedge.net/api/Indicator"
indicators_response = requests.get(indicators_url)

if indicators_response.status_code == 200:
    indicators_data = indicators_response.json()['value']
else:
    print("Failed to fetch indicators data")

# Create a dictionary of indicator codes and their descriptions
indicator_descriptions = {item['IndicatorCode']: item['IndicatorName'] for item in indicators_data}

# List of desired indicators
desired_indicators = [
    'Current health expenditure (CHE) as percentage of gross domestic product (GDP) (%)',
    'Healthy life expectancy (HALE) at birth (years)',
    'Maternal mortality ratio (per 100 000 live births)',
    'Age-standardized suicide rates (per 100 000 population)',
    'Deaths per 1 000 live births',
    'Medical doctors (per 10,000)',
    'Premature deaths due to noncommunicable diseases (NCD) as a proportion of all NCD deaths',
    'Population with household spending on health greater than 10% of total household budget (SDG 3.8.2, estimated data) (millions)',
    'Underweight numbers among children under 5 years of age (millions), model-based estimates',
    'UHC Service Coverage sub-index on noncommunicable diseases'
]

# Find and print the codes for the desired indicators
desired_indicator_codes = {description: code for code, description in indicator_descriptions.items() if description in desired_indicators}

del indicator_descriptions
del desired_indicators
del indicators_data
del indicators_response
del indicators_url

print(desired_indicator_codes)
```

Now we fetch the data using the desired_indicator_codes

```python
import requests
import pandas as pd
import time
import warnings

# Suppress the FutureWarning for the append method
warnings.filterwarnings("ignore", category=FutureWarning)

# Define the base URL for the WHO GHO API with the specified date range
base_url = "https://ghoapi.azureedge.net/api/{}?$filter=SpatialDim eq '{}' and date(TimeDimensionBegin) ge 2010-01-01 and date(TimeDimensionBegin) lt 2023-01-01"

# Define the series codes for the WHO GHO API from last code

# Define the country codes
#country_codes = {
#    'Uruguay': 'URY',
#}

# Function to fetch data for a specific indicator and country
def fetch_data(country_code, indicator_code):
    url = base_url.format(indicator_code, country_code)
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json().get('value', [])
        return data
    else:
        print(f"Error fetching data for {country_code} - {indicator_code}: {response.status_code}")
        return []

# Create an empty DataFrame to store the results
who_gho_df_raw = pd.DataFrame()

# Fetch data for each country and indicator
for country_name, country_code in country_codes.items():
    for series_name, series_code in desired_indicator_codes.items():
        print(f"Downloading data for {country_name} - {series_name}")
        records = fetch_data(country_code, series_code)
        for record in records:
            who_gho_df_raw = who_gho_df_raw.append({
                "country": country_name,
                "country_code": country_code,
                "year": record['TimeDim'],
                "indicator": series_name,
                "indicator_code": series_code,
                "value": record.get('Value')
            }, ignore_index=True)
        #time.sleep(1)  # Short pause between requests to avoid overloading the server

# Rearrange columns to the specified order
who_gho_df_raw = who_gho_df_raw[["country", "country_code", "year", "indicator", "indicator_code", "value"]]

# Save the raw data to a CSV file
who_gho_df_raw.to_csv("../data/output/who_gho_raw.csv", index=False)

# Load the CSV file to reshape the DataFrame
who_gho_short = pd.read_csv("../data/output/who_gho_raw.csv")

# Reshape the DataFrame to have new columns for each indicator value
who_gho_df = who_gho_short.pivot_table(index=['country', 'country_code', 'year'], columns='indicator', values='value')

# Reset the index to flatten the DataFrame
who_gho_df.reset_index(inplace=True)

# Save the reshaped DataFrame to a new CSV file
who_gho_df.to_csv('../data/output/new_who_gho_df.csv', index=False)

# Order the DataFrame by year and indicator
who_gho_short = who_gho_short.sort_values(by=['year', 'indicator'])

# Apply empirical decisions

# *Deaths per 1 000 live births*
mortality_data = who_gho_short[who_gho_short['indicator'] == 'Deaths per 1 000 live births']
mortality_data['value'] = mortality_data['value'].apply(lambda x: float(str(x).split()[0]))
mortality_data['value'] = mortality_data.groupby(['year', 'country'])['value'].transform('sum')
mortality_data = mortality_data.drop_duplicates()

# *Age-standardized suicide rates (per 100 000 population)*
suicide_data = who_gho_short[who_gho_short['indicator'] == 'Age-standardized suicide rates (per 100 000 population)']
suicide_data['value'] = suicide_data['value'].apply(lambda x: float(str(x).split()[0]))
suicide_data = suicide_data.groupby(['year', 'country', "country_code","indicator","indicator_code"])['value'].median().reset_index()

# *Premature deaths due to noncommunicable diseases (NCD)*
ncd_deaths_data = who_gho_short[who_gho_short['indicator'] == 'Premature deaths due to noncommunicable diseases (NCD) as a proportion of all NCD deaths']
ncd_deaths_data = ncd_deaths_data.groupby(['year', 'country']).nth(1).reset_index()  # Take the middle value
ncd_deaths_data['value'] = ncd_deaths_data['value'].apply(lambda x: float(str(x).split()[0])) 

# *Maternal mortality ratio (per 100 000 live births)*
maternal_mortality_data = who_gho_short[who_gho_short['indicator'] == 'Maternal mortality ratio (per 100 000 live births)']
maternal_mortality_data['value'] = maternal_mortality_data['value'].apply(lambda x: float(str(x).split()[0].replace(',', '')))


# Combine processed data
who_processed = pd.concat([
    mortality_data,
    suicide_data,
    ncd_deaths_data,
    who_gho_short[who_gho_short['indicator'].isin([
        'Current health expenditure (CHE) as percentage of gross domestic product (GDP) (%)',
        'UHC Service Coverage sub-index on noncommunicable diseases',
        'Medical doctors (per 10,000)'
    ])],
    maternal_mortality_data
])

# Save the processed data to a new CSV file
who_processed.to_csv("../data/output/processed_who_gho_data.csv", index=False)

# Reshape the DataFrame to have new columns for each indicator value
who_gho_df = who_processed.pivot_table(index=['country', 'country_code', 'year'], columns='indicator', values='value')

# Reset the index to flatten the DataFrame
who_gho_df.reset_index(inplace=True)

# Display the reshaped DataFrame
print(who_gho_df.head(40))

# Save the reshaped DataFrame to a new CSV file
who_gho_df.to_csv('../data/output/who_gho_df.csv', index=False)

del who_processed
del mortality_data
del suicide_data
del maternal_mortality_data
del ncd_deaths_data

```

The request brings more data than expected. We take the following empirical decisions. Do this after ordering the dataframe who_gho_short by year, by indicator.

- *Deaths per 1 000 live births
Add up values per year, so that we have Death from 0-4 years of age.

- *Age-standardized suicide rates (per 100 000 population)
Take middle value for each year (there are three values per year, take the value number two), for each country, as others are IC bounds. Then take only initial part of value, which is a number.

- *Premature deaths due to noncommunicable diseases (NCD) ...
Take middle value for each year (there are three values per year, take the value number two), for each country, as others are IC bounds. Then take only initial part of value, which is a number.

- *Current health expenditure (CHE) as percentage GDP (%)
None

- *UHC Service Coverage sub-index on noncommunicable diseases
None

- *Medical doctors (per 10,000)
None

- *Maternal mortality ratio (per 100 000 live births)
Take only the value as int, not the IC

- *Healthy life expectancy (HALE) at birth (years)
Take only the value in the middle for each year (there are three values per year, take the value number two)

```python
import pandas as pd
import numpy as np

def transform_dataframe(file_path, suffix):
    # Load the dataset
    data = pd.read_csv(file_path)

    # Transform the dataset to panel data structure
    df_transformed = data.melt(id_vars=['iso', 'Country/Region', 'unit'],
                               var_name='year',
                               value_name=f'ghg_{suffix}')

    # Replace any non-float or non-string values with np.nan
    df_transformed[f'ghg_{suffix}'] = pd.to_numeric(df_transformed[f'ghg_{suffix}'], errors='coerce')

    # Rearrange columns to match the required order
    df_transformed = df_transformed[['iso', 'Country/Region', 'year', f'ghg_{suffix}']]

    # Rename columns to match the desired format
    df_transformed = df_transformed.rename(columns={
        'iso': 'country_code',
        'Country/Region': 'country',
        'year': 'year'
    })

    return df_transformed

# List of file paths and corresponding suffixes
file_paths = [
    ('../data/input/climatewatch_agro_pc.csv', 'agro_pc'),
    ('../data/input/climatewatch_agro.csv', 'agro'),
    ('../data/input/climatewatch_total_pc.csv', 'total_pc'),
    ('../data/input/climatewatch_total.csv', 'total'),
    ('../data/input/climatewatch_trans_pc.csv', 'trans_pc'),
    ('../data/input/climatewatch_trans.csv', 'trans'),
    ('../data/input/climatewatch_agro_pgdp.csv', 'agro_pgdp'),
    ('../data/input/climatewatch_total_pgdp.csv', 'total_pgdp'),
    ('../data/input/climatewatch_trans_pgdp.csv', 'trans_pgdp'),
    ('../data/input/climatewatch_trans_co2_total.csv','co2_tr_total'),
    ('../data/input/climatewatch_trans_co2_pc.csv','co2_tr_pc'),
    ('../data/input/climatewatch_trans_co2_pgdp.csv','co2_tr_pgdp'),
    ('../data/input/climatewatch_co2_total.csv','co2_total_em'),
    ('../data/input/climatewatch_ind_co2_total.csv','co2_ind_total'),
    ('../data/input/climatewatch_ind_co2_pc.csv','co2_ind_pc'),
    ('../data/input/climatewatch_ind_co2_pgdp.csv','co2_ind_pgdp')
]

# Initialize an empty DataFrame for the final merged result
cw_new_df = None

# Process each file and merge the results
for file_path, suffix in file_paths:
    df = transform_dataframe(file_path, suffix)
    
    if cw_new_df is None:
        cw_new_df = df
    else:
        cw_new_df = pd.merge(cw_new_df, df, on=['country_code', 'country', 'year'], how='outer')

# Convert all columns except 'country' and 'country_code' to floats
columns_to_convert = cw_new_df.columns.difference(['country', 'country_code'])
cw_new_df[columns_to_convert] = cw_new_df[columns_to_convert].astype(float)

# Filter the dataframe for years between 2010 and 2023
cw_new_df = cw_new_df.loc[(cw_new_df["year"] > 2009) & (cw_new_df["year"] < 2024), ]

cw_new_df["ghg_agro_sh"] = cw_new_df["ghg_agro"] / cw_new_df["ghg_total"]
cw_new_df["ghg_trans_sh"] = cw_new_df["ghg_trans"] / cw_new_df["ghg_total"]
cw_new_df["ghg_co2_tr_sh"] = cw_new_df["ghg_co2_tr_total"] / cw_new_df["ghg_co2_total_em"]
cw_new_df["ghg_co2_in_sh"] = cw_new_df["ghg_co2_ind_total"] / cw_new_df["ghg_co2_total_em"]

cw_new_df = cw_new_df.loc[cw_new_df["ghg_agro_sh"]>0,]
cw_new_df = cw_new_df.loc[cw_new_df["ghg_trans_sh"]>0,]
cw_new_df = cw_new_df.loc[cw_new_df["ghg_co2_in_sh"]>0,]
cw_new_df = cw_new_df.loc[cw_new_df["ghg_co2_tr_sh"]>0,]

# Save the final merged dataframe as a CSV file
output_file_path = '../data/output/cw_new_df.csv'  # Adjust path as needed
cw_new_df.to_csv(output_file_path, index=False) 
```
